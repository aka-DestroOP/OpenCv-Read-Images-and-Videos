```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 as cv
from google.colab.patches import cv2_imshow
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
with_mask_files=os.listdir('/content/data/with_mask')
print(with_mask_files[0:5]) #selecting first 5 elements
print(with_mask_files[-5:]) #selecting last 5 elements
without_mask_files=os.listdir('/content/data/without_mask')
print(without_mask_files[0:5]) #selecting first 5 elements
print(without_mask_files[-5:]) #selecting last 5 elements
```

```python
print("No. of with mask images",len(with_mask_files))
print("No. of without mask images",len(without_mask_files))
```

**creating labels for two class of images**


with mask    -> 1

without mask -> 0

```python
#creating the labels
with_mask_labels=[1]*3725
without_mask_labels=[0]*3828
print(with_mask_labels[0:5])
print(without_mask_labels[0:5])
print(len(with_mask_labels))
print(len(without_mask_labels))
```

```python
labels=with_mask_labels+without_mask_labels
print(len(labels))
print(labels[0:5])
print(labels[-5:])
```

**Displaying the images**

```python
img=mpimg.imread('/content/data/with_mask/with_mask_1003.jpg')
imgplot=plt.imshow(img)
plt.show()
```

```python
img=mpimg.imread('/content/data/without_mask/without_mask_700.jpg')
imgplot=plt.imshow(img)
plt.show()
```

**Image Processing**

1. resize the images

2. convert the images into numpy arrays

```python
data=[]
with_mask_path='/content/data/with_mask/'
for img_file in with_mask_files:
  image=Image.open(with_mask_path+img_file)
  image=image.resize((128,128))
  image=image.convert('RGB')      #not all images are in rgb
  image=np.array(image)
  data.append(image)

without_mask_path='/content/data/without_mask/'
for img_file in without_mask_files:
  image=Image.open(without_mask_path+img_file)
  image=image.resize((128,128))
  image=image.convert('RGB')
  image=np.array(image)
  data.append(image)
```

```python
data[0]
```

```python
print(data[0])
type(data[0])
```

```python
#converting image list and label list into numpy arrays
x=np.array(data)
y=np.array(labels)
print(x.shape)
print(y.shape)
print(y)
```

**train test split**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

**scaling**

```python
x_train_scale=x_train/255
x_test_scale=x_test/255
```

```python
x_train[0]
```

```python
print(x_train_scale[0])
```

**Building a Convolutional Neural Network(CNN)**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
num_of_classes=2
model=keras.Sequential()
model.add(keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(128,128,3)))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Conv2D(64,(3,3),activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Flatten())

model.add(keras.layers.Dense(128,activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add((keras.layers.Dense(num_of_classes,activation='sigmoid')))
```

sigmoid is used for binary class classification


softmax is used for multi-class classification

**Compiling the Neural Network**

1. Optimisation Algorithm

2. Loss function matrix

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
```

```python
history=model.fit(x_train_scale,y_train,validation_split=0.1,epochs=5)
```

**Model Evaluation**

```python
loss,accuracy=model.evaluate(x_test_scale,y_test)
print('Test Accuracy =',accuracy)
```

```python
h=history
#plotting the loss value
plt.plot(h.history['loss'],label='train loss')
plt.plot(h.history['val_loss'],label='validation loss')
plt.legend()
plt.show()

#plotting the accuracy value
plt.plot(h.history['acc'],label='train acurracy')
plt.plot(h.history['val_acc'],label='validation acurracy')
plt.legend()
plt.show()
```

**Predictive System**

```python
input_image_path=input('Path of the image to be predicted')
input_image=cv.imread(input_image_path)
cv2_imshow(input_image)
input_image_resize=cv.resize(input_image,(128,128))
input_image_scale=input_image_resize/255
input_image_reshape=np.reshape(input_image_scale,[1,128,128,3])
prediction=model.predict(input_image_reshape)
print(prediction)
prediction_label=np.argmax(prediction)
print(prediction_label)
if prediction_label==0:
  print('The person in the image is not wearing mask')
else:
  print('The person in the image is wearing mask')
```
